# Multivariate normal PDF

$$
\text{MVNormalPDF}(x | \mu, \Sigma) = \frac{1}{(2\pi)^{D/2}} \frac{1}{\text{det}(\Sigma)^{1/2}} \exp{-\frac{1}{2} (x-\mu)^T \Sigma^{-1} (x-\mu) }
$$
